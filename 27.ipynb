```python
path = '/content/drive/MyDrive/TGMT'
os.listdir(path)
```

```python
img = cv2.imread(path+'/anh01.jpg')
```

```python
cv2_imshow(img)
```

```python
# Ti m hnh  c hun luyn t trc
net = cv2.dnn.readNetFromCaffe('/content/drive/MyDrive/TGMT/Model/deploy.prototxt.txt', '/content/drive/MyDrive/TGMT/Model/res10_300x300_ssd_iter_140000_fp16.caffemodel')

# Chun b d liu u vo
# img: y l nh u vo m bn mun nhn dng khun mt. Trong trng hp ca bn, bn  c nh t tp "face.png".
# 1.0: y l t l co gin cho nh. Trong trng hp ny, nh s khng b co gin hoc m rng, v gi nguyn kch thc ban u.
# (300, 300): y l kch thc m m hnh yu cu cho nh u vo. M hnh m bn ang s dng mong mun nh c kch thc 300x300 pixel. Do , bn co gin hoc ct nh u vo thnh kch thc ny.
# (104, 177, 123): y l gi tr trung bnh mu sc c tr i t mi pixel ca nh. iu ny thng c s dng  chun ha d liu u vo. Trong trng hp ny, cc gi tr ny thng c ly t d liu hun luyn ca m hnh.
# swapRB=False: y l mt c  xc nh xem c cn hon i cc knh mu  v xanh (Red-Blue) trong nh hay khng. Trong trng hp ny, bn  t n thnh False, tc l khng hon i knh m
blob = cv2.dnn.blobFromImage(img, 1.0, (300, 300), (104, 177, 123), swapRB=False)

# t d liu u vo cho mng
net.setInput(blob)

# Chy mng  pht hin khun mt
faces = net.forward()

# Ly kch thc ca nh u vo
h = img.shape[0]
w = img.shape[1]

# In thng tin
print(faces.shape)
print(faces[0, 0, 0, ])

# Duyt tng khun mt  c pht hin
for i in range(0, faces.shape[2]):
    confidence = faces[0,0,i,2]
    # Kim tra nu mt c  tin cy l 0.5
    if confidence>0.5:
        # Trch xut ta 
        print(faces[0, 0, i, 3:7])
        startx = int(faces[0, 0, i, 3]*w)
        starty = int(faces[0, 0, i, 4]*h)
        endx = int(faces[0, 0, i, 5]*w)
        endy = int(faces[0, 0, i, 6]*h)
        print(startx, starty, endx, endy)

        # V hnh ch nht xung quanh khun mt  pht hin
        cv2.rectangle(img, (startx, starty), (endx, endy), (0, 255, 0),)

        # Hin th  tin cy
        text = 'Face: {:.2f}%'.format(confidence*100);
        cv2.putText(img, text, (startx, starty-10), cv2.FONT_HERSHEY_PLAIN, 1, (255,255,255),)

# Hin th nh gc
cv2_imshow(img)
```
