```python
# import thư viện 
from PIL import Image
from imgtools import *
```

```python
# Đường dẫn ảnh
my_path = 'D:/anh/01.png'
```

```python
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100, 150, 200, 250)
# cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# xác định vị trí dán
paste_position = (150, 250)
img.paste(cropped_image, paste_position)
display(img)
```
