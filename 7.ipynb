```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Đường dẫn ảnh
my_path = 'D:/anh/01.png'
# Đọc ảnh và hiển thị
im = Image.open(my_path)
# display(im)
plt.imshow(im)
```

```python
# Vẽ trên hình ảnh
# ĐỌc ảnh
im = Image.open(my_path)
x = [100, 100, 400, 400]
y = [200, 300, 200, 300]

plt.imshow(im) # Hiển thị hình ảnh
plt.plot(x,y, 'bo--') 
plt.show()
```
